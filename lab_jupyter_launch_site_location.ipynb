```python
# TASK 3: Calculate the distances between a launch site to its proximities
1. Import Haversine Function
To calculate distance between two sets of coordinates (lat/lon), we can use the Haversine formula :
```

```python
from math import radians, sin, cos, sqrt, atan2

# Function to calculate distance between two points on Earth (in km)
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # Radius of Earth in kilometers
    
    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    # Differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    return R * c
```

```python
2. Define Proximity Points (Example: Coastline near CCAFS LC-40)
Lets pick one example point: the nearest coastline to CCAFS LC-40 , located at approximate coordinates:
```

```python
coastline_coords = [28.5636, -80.5716]  # Example: Closest coast near Cape Canaveral
railway_coords = [28.5721, -80.5677]   # Example: Nearby railway line
highway_coords = [28.5632, -80.5708]   # Example: Nearby highway (A1A)
city_coords = [28.4093, -80.6040]      # Example: Titusville, FL
```

```python
Add Distance Columns to DataFrame
Now calculate the distances from each launch to those points:
```

```python
# Get unique launch sites
launch_sites = spacex_df[['Launch Site', 'Lat', 'Long']].drop_duplicates().reset_index(drop=True)

# Create new columns for distances
launch_sites['distance_to_coast'] = 0.0
launch_sites['distance_to_railway'] = 0.0
launch_sites['distance_to_highway'] = 0.0
launch_sites['distance_to_city'] = 0.0

# Fill in distances
for index, row in launch_sites.iterrows():
    lat, lon = row['Lat'], row['Long']
    
    launch_sites.at[index, 'distance_to_coast'] = haversine_distance(lat, lon, *coastline_coords)
    launch_sites.at[index, 'distance_to_railway'] = haversine_distance(lat, lon, *railway_coords)
    launch_sites.at[index, 'distance_to_highway'] = haversine_distance(lat, lon, *highway_coords)
    launch_sites.at[index, 'distance_to_city'] = haversine_distance(lat, lon, *city_coords)

# Display result
launch_sites
```

```python
from math import radians, sin, cos, sqrt, atan2

def calculate_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the distance in kilometers between two points using Haversine formula.
    """
    R = 6371.0  # Earth radius in kilometers
    
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    
    return R * c
```
